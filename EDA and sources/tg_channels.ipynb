```python
DATA_PATH = "../data/"

# Replace with your Telegram API credentials
API_ID = ""
API_HASH = ""
```

```python
import pandas as pd

df = pd.read_parquet(DATA_PATH + 'train.parquet')
test_df = pd.read_csv(DATA_PATH + 'test.csv')

df.head()
```

```python
all_content = " ".join(df.content.values) + " ".join(test_df.content.values)

len(all_content)
```

# Tags

```python
import re

all_ats = list(set(re.findall(r'@[.a-zA-Z0-9_-]+', all_content)))
all_tags = [tag for tag in all_ats if "." not in tag]

all_tags
```

```python
len(all_tags)
```

```python
from telethon import TelegramClient
import asyncio
```

```python
import time

async def check_if_channel(client, username):
    time.sleep(0.5)
    try:
        entity = await client.get_entity(username)
        if entity.broadcast:  # "broadcast" is True for channels
            print(f"{username} is a Telegram channel.")
            return username, True
        else:
            print(f"{username} is NOT a channel.")
            return username, False
    except Exception as e:
        print(f"Error checking {username}: {e}")
        return username, None

async def main(usernames):
    async with TelegramClient("session", API_ID, API_HASH) as client:
        tasks = [check_if_channel(client, username.replace("t.me/", "").replace("@", "")) for username in usernames]
        results = await asyncio.gather(*tasks)
        return results
```

```python
result = await main(all_tags)
```

```python
channel_tags = []
for tag in result:
    if tag[1]:
        channel_tags.append(tag[0])
```

```python
channel_tags
```

```python
len(channel_tags)
```

```python
# open file
with open('test_tg_channels.txt', 'w+') as f:

    # write elements of list
    for items in channel_tags:
        f.write('%s\n' %items)

    print("File written successfully")


# close the file
f.close()
```

# Links

```python
regex = r'(http|ftp|https):\/\/([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:\/~+#-]*[\w@?^=%&\/~+#-])'

all_links = list(set(re.findall(regex, all_content)))
```

```python
all_links
```

```python
links_df = pd.DataFrame(all_links, columns=['http', 'domain', 'url'])

links_df['domain'].value_counts()
```

```python
links_df[links_df['domain'] == 't.me']
```

```python
tg_mask = links_df['domain'] == 't.me'

tags_from_links = links_df.loc[tg_mask, 'url'].apply(lambda x: x.split('/')[1]).tolist()

tags_from_links
```

```python
links_result = await main(tags_from_links)
```

```python
links_channel_tags = []
for tag in links_result:
    if tag[1]:
        links_channel_tags.append(tag[0])
```
